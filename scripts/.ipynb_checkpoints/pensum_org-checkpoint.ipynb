```python
DI = pensum[['Clave', 'Nombre', 'CR', 'PRE-REQ', 'Cuatrimestre']]
my_courses = pensum[['Clave', 'Nombre','PRE-REQ', 'Completed', 'Mes', 'Score']]
DI.head()
```

```python
conn = mysql.connector.connect(
    host="localhost",
    user="oa",  
    password="",
    database="pensum_db"

)

cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS program (
    code VARCHAR(20) PRIMARY KEY,
    name VARCHAR(100),
    credit INT,
    req VARCHAR(20)
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS courses (
    code VARCHAR(20) PRIMARY KEY,
    name VARCHAR(100),
    req VARCHAR(20),
    completed BOOLEAN,
    month VARCHAR(10),
    score FLOAT
)
""")
```

```python
cursor.execute("TRUNCATE TABLE program")
conn.commit()



DI = pensum[['Clave', 'Nombre', 'CR', 'PRE-REQ', 'Cuatrimestre']]


program_query = """
INSERT INTO program (code, name, credit, req, quarter)
VALUES (%s, %s, %s, %s, %s)
"""

for _, row in DI.iterrows():
    try:
        credit = int(row['CR'])
    except ValueError:
        print(f"Cannot convert credit to int for row: {row['Clave']} - value: {row['CR']}")
        credit = None  # or 0, or some default
    cursor.execute(
        program_query,
        (
            row['Clave'],
            row['Nombre'],
            credit,
            row['PRE-REQ'],
            row['Cuatrimestre']
        )
    )
conn.commit()



conn.commit()


print("✅ Data inserted into MySQL (no SQLAlchemy needed!)")
```

```python
my_courses['Mes'] = my_courses['Mes'].astype(str)



courses_query = """
INSERT INTO courses (code, name,req, completed, month, score)
VALUES (%s, %s, %s, %s, %s, %s)
"""

for _, row in my_courses.iterrows():
    cursor.execute(courses_query, tuple(row))

conn.commit()

```
